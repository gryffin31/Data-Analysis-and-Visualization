# 1) Analysing "rate" columns since ther are lot of null values present

```python
zomato_df['rate'].unique()
```

----> As we see that there are some faulty values in rating column = "NEW", "-" and replace them with "nan = i.e. by  null values" 

Since "rate" columns goves information about that persons views on that restaurants.

But these values does not give, so we have to remove them.

```python
zomato_df['rate']
```

```python
# Defining a function to handel the rating column and also covert them from text to integer form
def treat_ratings(values):
    if (values == 'NEW' or values == '-'):
        return np.nan                   ### NumPy NAN stands for not a number and 
                                   #    np.nan is defined as a substitute for declaring value which are numerical values that are missing values in an array as NumPy is used to deal with arrays in Python
    
    else :
        values = str(values).split('/')                # since split method splits string into list of int 
                                                       # i.e."4.1/5" will split it into 4.1 and /5 ---> list is =   values=[4.1][/5]
        values = values[0]                                # But here we only need numerator of rating because all the rating are out of 5
                                                       # So we have stated its index value.
        return float(values)                           # Since we want these values in floating form.
        
        
# Now applying the function on "rate" column        

zomato_df.rate = zomato_df['rate'].apply(treat_ratings)
    
```

```python
zomato_df['rate'].unique()
```

---> Now by observing above rate column after treating faulty values we get good view of ratings of restaurants.

Now "rate" column is converted from text to folat

```python
zomato_df.info()
```

```python
zomato_df.isnull().sum()            
```

Since Null Values of "Rate" feature increased from previous it is becoz we have filled faulty values by 'nan' values

 Since null values from "rate" column is quite high so we cant directly remove them becoz it will causes accuracy of model
So we will replace them by there "mean" -----> bcoz it is numerical feature so we have to replace  them by mean.

```python
## Using fill na method 

zomato_df.rate.fillna(zomato_df.rate.mean(), inplace= True)
```

```python
zomato_df.isnull().sum()
```

-----> So rate column has now no null values.

---> Now as compare to size of dataset in remaining features like ["location", "rest_type", "cuisines", "approx_cost(for two people)"] has very low null values present so if we drop them directly it wont affect our analysis much.

# 5) Now analysing cost of plate per cuisines in restaurant i.e. 

## "approx_cost(for two people) feature

Since approx_cost column is object datatype but it is actually a numeric variable so firstly we have to convert it.

```python
zomato_df['approx_cost(for two people)'].value_counts()
```

```python
zomato_df['approx_cost(for two people)'].unique()
```

Observation= 

1) as we see  that data type of cost feature is object becoz of comas in between number when approx cost is above 999. i.e. 1,900 so on 

2) so we have to remove these comas i.e. ","

```python
# Writing functing to remove "," and convert this column into float

def handling_approx_cost(cost):
    cost = str(cost)
    if "," in cost:
        cost = cost.replace("," , "")           # we simply replaced coma with empty value
        return float(cost)                      # converting data type into float
    else:
        return float(cost)
    
# applying fun on feature
zomato_df['approx_cost(for two people)'] = zomato_df['approx_cost(for two people)'].apply(handling_approx_cost)
```
